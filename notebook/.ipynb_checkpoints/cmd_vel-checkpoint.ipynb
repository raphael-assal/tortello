```python
def prev_filter_dec(t, var=.2):
    prev = t
    def prev_filter(t):
        nonlocal prev
        if abs((prev - t) / prev) > var:
            print(f"prev={prev}, t={t}, filtered-out !")
            return prev
        else:
            prev = t
            print(f"prev={prev}, t={t}, pass !")
            return t
    return prev_filter
        

```

```python
f = prev_filter_dec(33)
```

```python
f = prev_filter_dec(odom.x[0])
x = odom.x.values.tolist()
```

```python
x_filt = [f(x) for x in x]
```

```python
plt.plot(x)
plt.plot(x_filt)
```
